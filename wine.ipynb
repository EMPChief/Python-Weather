## Did some simple import and loading data


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('wines.csv')
```

```python
dataframe
```

## Checking how many wines that get 100points


```python
point_check = len(dataframe[dataframe['points'] == 100])
print(point_check)
```

## What is the price off most expensive wine in dataset


```python
expensive = dataframe.loc[dataframe['price'] ==
                          dataframe['price'].max()]["name"].squeeze()
print(expensive)
```

## Calculate new column where check the rating


```python
dataframe["rate"] = dataframe["points"] / 20
dataframe
```

## Create a histagram for those cost less then 100


```python
plt.hist(dataframe.loc[dataframe["price"] < 100]
         ["price"], bins=10, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Wine Prices (Less than $100)')
plt.show()
```

## Plot price horizontally against points vertically to check if price increased with rating


```python
dataframe.plot(x='price', y="points", figsize=(15, 3), kind="scatter")
plt.xlabel('Price')
plt.ylabel('Points')
plt.title('Scatter Plot of Wine Prices vs Points')
plt.grid(True)
plt.show()
```
